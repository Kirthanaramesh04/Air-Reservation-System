```python
import pymongo
import re  
def add_passenger_to_mongodb():
    client = pymongo.MongoClient("mongodb://localhost:27017/")
    db = client["class_db"]
    collection = db["passenger"]
    print("!!!WELCOME TO THE AIR RESERVATION SYSTEM!!!")
    user_input = input("Enter 'yes' if you are a new user, otherwise enter 'no': ").lower()
    
    if user_input == "no":
        login_info(collection)
    else:
        registration_info(collection)

    reservation_info(collection)
    price_info(collection)
    transaction_info(collection)
    airport_info(collection)
    cancel_option = input("Do you want to cancel your reservation? (yes/no): ").lower()
    if cancel_option == "yes":
        cancel_reservation(collection)
    
    client.close()

def login_info(collection):
    print("!!! LOGIN DETAILS !!!")
    customer_id = int(input("Enter the customer id: "))
    password = get_validated_password()
    login = {
        "Customer_id": customer_id,
        "Password": password,
    }
    collection.insert_one(login)
    print("Login Successful!")

def registration_info(collection):
    print("!!! REGISTRATION !!!")
    first_name = input("Enter the First Name: ")
    last_name = input("Enter the Last Name: ")
    address = input("Enter the address: ")
    contact_no = get_validated_phone_number()
    city = input("Enter the city: ")
    state = input("Enter the state: ")
    country = input("Enter the country: ")
    email_id = get_validated_email()
    customer_id = int(input("Enter the id: "))
    password = get_validated_password()
    passenger_data = {
        "FirstName": first_name,
        "LastName": last_name,
        "Address": address,
        "Contact_no": contact_no,
        "City": city,
        "State": state,
        "Country": country,
        "Email_id": email_id,
        "customer_id": customer_id,
        "password": password,
    }
    collection.insert_one(passenger_data)
    print("Registration successful!")

def get_validated_phone_number():
    while True:
        contact_no = input("Enter the contact number (10 digits): ")
        if len(contact_no) == 10 and contact_no.isdigit():
            return contact_no
        else:
            print("Invalid phone number. Please enter a 10-digit number.")

def get_validated_email():
    while True:
        email_id = input("Enter the email_id: ")
        if re.match(r"[^@]+@[^@]+\.[^@]+", email_id):
            return email_id
        else:
            print("Invalid email format. Please enter a valid email address.")

def get_validated_password():
    while True:
        password = input("Enter a password (minimum 5 characters, including one special character): ")
        if len(password) >= 5 and any(char.isdigit() or char.isalpha() or not char.isalnum() for char in password):
            return password
        else:
            print("Invalid password. Please meet the specified criteria.")

def reservation_info(collection):
    print("!!! FLIGHT RESERVATION !!!")
    flight_num = int(input("Enter the Flight number: "))
    flight_name = input("Enter the Flight name: ")
    dep_time = input("Enter a Departure time: ")
    arr_time = input("Enter an Arrival time: ")
    origin = input("Enter an origin: ")
    destination = input("Enter the destination: ")
    reservation_data = {
        "Flight Number": flight_num,
        "Flight_Name": flight_name,
        "Departure_Time": dep_time,
        "Arrival_Time": arr_time,
        "Origin": origin,
        "Destination": destination,
        
    }
    collection.insert_one(reservation_data)
    print("Flight Details noted Successfully!")

def price_info(collection):
    print("!!! PRICE INFO !!!")
    print("Enter the type of class: ")
    print("1. First")
    print("2. Second")
    print("3. Third")
    class_choice = input("Enter the option: ")
    
   
    if class_choice not in ['1', '2', '3']:
        print("ENTER THE RIGHT OPTION")
        return

    num_of_seats = int(input("Enter the number of seats: "))

   
    price_per_seat = calculate_price_per_seat(class_choice)

    if price_per_seat is not None:
        total_price = num_of_seats * price_per_seat
        print(f"Price per seat: {price_per_seat}")
        print(f"Total Price for {num_of_seats} seat(s) is {total_price}")
        price_data = {
            "class": class_choice,
            "num_of_seats": num_of_seats,
            "total_price": total_price
        }
        collection.insert_one(price_data)
    else:
        print("Invalid class choice. Unable to calculate price.")

def calculate_price_per_seat(class_choice):
    class_prices = {'1': 20000, '2': 15000, '3': 10000}
    return class_prices.get(class_choice)

def transaction_info(collection):
    print("!!! PAYMENT DETAILS !!!")
    payment_type = input("Enter (C or c) for CREDIT CARD, (D or d) for DEBIT CARD, or (G or g) for UPI: ")
    
    if payment_type.lower() == 'c':
        credit_num = get_validated_credit_card()
        pin_num = get_validated_pin("credit card")
    elif payment_type.lower() == 'd':
        debit_num = get_validated_debit_card()
        pin_num = get_validated_pin("debit card")
    elif payment_type.lower() == 'g':
        upi_id = input("Enter your UPI ID: ")
        pwd = get_validated_pin("UPI")
    else:
        print("Enter the Right Value: ")
        return

    transaction_data = {
        "Payment_type": payment_type,
    }
    collection.insert_one(transaction_data)
    print("Payment successful!")

def get_validated_credit_card():
    while True:
        credit_card_num = input("Enter credit card number (16 digits): ")
        if len(credit_card_num) == 16 and credit_card_num.isdigit():
            return credit_card_num
        else:
            print("Invalid credit card number. Please enter a 16-digit number.")
            
def get_validated_debit_card():
    while True:
        debit_card_num = input("Enter debit card number (16 digits): ")
        if len(debit_card_num) == 16 and debit_card_num.isdigit():
            return debit_card_num
        else:
            print("Invalid debit card number. Please enter a 16-digit number.")

def get_validated_pin(card_type):
    while True:
        pin_num = input(f"Enter {card_type} pin number (4 digits): ")
        if len(pin_num) == 4 and pin_num.isdigit():
            return pin_num
        else:
            print(f"Invalid {card_type} pin number. Please enter a 4-digit number.")

def airport_info(collection):
    print("!!! AIRPORT DETAILS !!!")
    country_name = input("Enter your country name: ")
    airport_name = input("Enter the airport name: ")
    airport_data = {
        "CountryName": country_name,
        "AirportName": airport_name,
    }
    collection.insert_one(airport_data)
def cancel_reservation(collection):
    print("!!! CANCEL FLIGHT RESERVATION !!!")
    customer_id = int(input("Enter your customer ID: "))
    confirmation = input("Are you sure you want to cancel your reservation? (yes/no): ").lower()
    if confirmation == "yes":
        result = collection.delete_one({"Customer_id": customer_id})
        if result.deleted_count == 1:
            print("Reservation successfully canceled.")
        else:
            print("No reservation found for the provided customer ID.")
    else:
        print("Reservation cancellation aborted.")    
    

add_passenger_to_mongodb()
print("* YOUR FLIGHT HAS BEEN BOOKED. HAPPY JOURNEY *")
```
